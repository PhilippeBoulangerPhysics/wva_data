```python
def mask_p_under_ps(pfull,ps):
    mask = pfull[np.newaxis,:,np.newaxis,np.newaxis] <= ps[:,np.newaxis,:,:]
    return mask
g = 9.81

def vertical_integral(field,pfull,ps,dp):
    mask = mask_p_under_ps(pfull,ps)
    masked_field = mask*field
    tmp = masked_field * dp[np.newaxis,:,np.newaxis,np.newaxis]/g
    integral = np.sum(tmp,axis = 1)
    return integral

from scipy.interpolate import interp1d
def get_interpolated_baseline_field_and_field(field,lat):
    equivalent_lat = lat.copy()
    lat_interp = np.linspace(equivalent_lat[0],equivalent_lat[-1],len(equivalent_lat)*2)
    f = interp1d(lat, field, axis=1, kind='linear')
    field_interp = f(lat_interp)
    print(field_interp.shape)
    field_zonal = np.mean(field_interp,axis = 2)
    field_zonal_time = np.mean(field_zonal,axis = 0)

    return field_zonal_time, field_interp,lat_interp

def find_equivalent_latitude():
    pass

```
